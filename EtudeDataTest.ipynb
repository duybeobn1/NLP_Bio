```python
import pandas as pd
from collections import Counter

# Lire les fichiers CSV / TXT
train = pd.read_csv("/Users/bgdu69/Documents/Hackermann/Bio Inspired ML/NLP_Bio/dataset/train.txt", sep=";") 
test = pd.read_csv("/Users/bgdu69/Documents/Hackermann/Bio Inspired ML/NLP_Bio/dataset/test.txt", sep=";")
val = pd.read_csv("/Users/bgdu69/Documents/Hackermann/Bio Inspired ML/NLP_Bio/dataset/val.txt", sep=";")

# VÃ©rification rapide
print("Train :", train.shape)
print("Test  :", test.shape)
print("Val   :", val.shape)

datasets = {"train": train, "val": val, "test": test}
kk = [15999, 1999, 1999]
i=0

for name, df in datasets.items():
    print(f"\nğŸ“Š Distribution des classes pour {name} :")
    counts = df.iloc[:, 1].value_counts()  # la 2e colonne = labels
    print(counts/kk[i])
    i += 1

# --- Fonction simple de tokenisation ---
def tokenizer(text):
    return text.lower().split()

# --- Fonction pour compter les mots rares ---
def rare_words_count(texts, threshold=2, tokenizer=None):
    if tokenizer is None:
        tokenizer = lambda x: x.lower().split()
    
    all_tokens = [token for text in texts for token in tokenizer(text)]
    token_counts = Counter(all_tokens)
    rare_words = [word for word, count in token_counts.items() if count < threshold]
    
    print(f"Nombre de mots rares (<{threshold} occurrences) : {len(rare_words)}")
    return rare_words, token_counts

# --- Parcourir les datasets ---
for name, df in datasets.items():
    print(f"\nğŸ“Š Distribution des classes pour {name} :")
    counts = df.iloc[:, 1].value_counts()
    proportions = counts / len(df)
    print(proportions)
    
    # Compter les mots rares dans la colonne de texte (colonne 0)
    print(f"\nğŸ“ Comptage des mots rares pour {name} :")
    rare_words, token_counts = rare_words_count(df.iloc[:, 0].tolist(), threshold=2, tokenizer=tokenizer)
```
